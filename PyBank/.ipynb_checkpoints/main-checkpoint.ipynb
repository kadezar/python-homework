```python
import os
import csv
import pandas as pd
from pathlib import Path
```

```python
budget_df = pd.read_csv(
    Path("budget_data.csv"),
    index_col = "Date"
)
budget_df
```

```python
previous_revenue = 0
month_of_change = []
revenue_change = 0
greatest_decrease = ["", 9999999]
greatest_increase = ["", 0]
revenue_change_list = []
revenue_average = 0

with open('budget_data.csv') as csvfile: 
    csvreader = csv.DictReader(csvfile)
    previous_revenue = float(next(csvreader)["Profit/Losses"])  
    for row in csvreader:
        
# Calculate total number of months
        total_months = len(budget_df["Profit/Losses"])
    
# Calculate net total amount
        net_total = budget_df["Profit/Losses"].sum()
    
# Calculate average of the changes in Profit/Losses
        revenue_change = float(row["Profit/Losses"])- previous_revenue
        previous_revenue = float(row["Profit/Losses"])
        revenue_change_list = revenue_change_list + [revenue_change]
        revenue_average = round(sum(revenue_change_list)/len(revenue_change_list), 2)
        month_of_change = [month_of_change] + [row["Date"]]
        
# Calculate greatest increase
        if revenue_change>greatest_increase[1]:
            greatest_increase[1]= revenue_change
            greatest_increase[0] = row['Date']

#The greatest decrease in revenue (date and amount) over the entire period
        if revenue_change<greatest_decrease[1]:
            greatest_decrease[1]= revenue_change
            greatest_decrease[0] = row['Date']
        

print(f"Total Months: ${total_months}")
print(f"Total: ${net_total}")
print(f"Average Change: ${revenue_average}")
print(f"Greates Increase in Profits: {greatest_increase[0], greatest_increase[1]}")
print(f"Greatest Decrease in Profits: {greatest_decrease[0], greatest_decrease[1]}")
```
